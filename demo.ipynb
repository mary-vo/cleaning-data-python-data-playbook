```python
# When trying to convert to numeric, there are values with string 'mm', coerce the error 
pd.to_numeric(fulldf.height, errors="coerce")
```

# Aggregating Data

```python
import pandas as pd
```

```python
data = pd.read_csv('artwork_sample.csv')
data.head()
```

```python
# We can aggregate year, by calling an aggregate method directly on the series
print(data.year) # returns some nan values

# nan are automatically excluded from aggregate functions
print(data.year.min())
print(data.year.max())
print(data.year.sum())
print(data.year.mean())
print(data.year.std())
```
