# Aggregating Data

```python
import pandas as pd
```

```python
data = pd.read_csv('artwork_sample.csv')
data.head()
```

```python
# We can aggregate year, by calling an aggregate method directly on the series
print(data.year) # returns some nan values

# nan are automatically excluded from aggregate functions
print(data.year.min())
print(data.year.max())
print(data.year.sum())
print(data.year.mean())
print(data.year.std())
```

```python
# You can also sum across the entire dataframe, it might not always make sense though
data.sum()
```

```python
# These do the same thing BUT with agg function you can pass multiple aggregation function sat the same time
print(data.agg('min'))
print(data.min())
# print(data.agg(['min','max','mean','std']))
```

# Normalizing Data

```python
import pandas as pd
```

```python
data = pd.read_csv('artwork_sample.csv')
data.head()
```

```python
# When normalizing data, you're lookiong at numeric data (float or int)
data.dtypes
```

```python
# normalize height
height = data.height

# Standardization
norm = (height - height.mean()) / height.std()
print(norm)

# Normalize all values between 0 and 1
minmax = (height - height.min()) / (height.max() - height.min())
print(minmax)
```

```python
# Nothing has changed in the original dataframe yet
data
```

```python
# Set the original height column to the normalized values
data.height = minmax

# OR create a bnand new column
data['standardized_height'] = norm
```

```python
data.head()
```

# Transforming Data

```python
import pandas as pd
```

```python
data = pd.read_csv('artwork_sample.csv')
```

```python
# Transform height column
data.height.transform(lambda x: print(x))
```

```python
data.height.transform(lambda x: x)
```

```python
data.height.transform(lambda x: x/10)
```

```python
# Transform with groupby
print(data.groupby('artist'))
data.groupby('artist').transform('nunique')
```

```python
print(data.groupby('artist')['height'])
print(data.groupby('artist')['height'].transform('mean'))
print(data.artist)
```

```python
# Create new column to add to dataframe (data) and ssign mean height by artist 
data['mean_height_by_artist'] = data.groupby('artist')['height'].transform('mean')
```

# Filtering Data

```python
import pandas as pd
```

```python
data = pd.read_csv('artwork_data.csv', low_memory = False)
```

```python
data.filter(items=['id','artist'])
```

```python
# like is case sensitive
print(data.filter(like='year'))

# use regex
print(data.filter(regex="(?i)year"))
```
