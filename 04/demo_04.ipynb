# Direct Filtering with Square Brackets

```python
import pandas as pd
```

```python
data = pd.read_csv(sample_path)
```

```python
# To see a specific column, this returns a series
print(data['id'])

# Returns row with index 1
print(data['id'][1])
```

```python
# For multiple columsn, pass an array into the square brackets
# This returns a DATAFRAME, cannot return based on index anymore
data[['artist','title']]
```

```python
# This will return a series of artist only and now we can access a single element from the series
data[['artist','title']]['artist'][1]
```

```python
# Although you cannot access a single row, you can get a range of rows
# Inclusive of first index and exclusive of end index
data[1:3]
```

```python
# This is still a dataframe, so we can still access an individual column with square brackets
data[1:5]['artist']
```

```python
data[data['year'] > 1800]
```

```python
data[data['year'] > 1800]['year'] # returns a series
```

# Data Indexing with .loc

```python
import pandas as pd
```

```python
data = pd.read_csv(sample_path)
```

```python
# data.loc[ROWS,COLS]
print(data.loc[0, :])
print(data.loc[0:2, :]) #range is inclusive of start and end
print(data[0:2]) #range is exlusive of the end
```

```python
data.loc[0:2, 'title'] # returns a series
```

```python
# Pass an array of columns for multiple columns,w hich returns a dataframe
data.loc[0:2, ['artist','title']]

# Pass an array of row labels
data.loc[[1,5],['artist','title']]
```

```python
data.head()
```

```python
data.loc[[1,5],'id':'artistId']
```

```python
data.loc[data.artist == 'Blake, Robert', :]
```

```python
data.loc[data.artist == 'Blake, Robert', ['title','year']]
```

# Using .iloc to Access Specific Rows or Columns

```python
import pandas as pd
```

```python
data = pd.read_csv(sample_path)
```

```python
# Return just the first row
data.iloc[0,:]
```

```python
# loc vs iloc
print(data.loc[0:3,:]) #inclusive
print(data.iloc[0:3, :]) #exclusive of the end
```

```python
# To better see difference of loc and iloc, set the index to id
data.set_index('id', inplace=True)
```

```python
# loc vs iloc
print(data.loc[1035:1037,:]) #inclusive
print(data.iloc[0:3, 0:3]) #exclusive of the end
```

```python
data.reset_index(inplace=True)
```

```python
data.iloc[[5,8],[0,2]]
```

# Filering with str.contains

```python
import pandas as pd
```

```python
data = pd.read_csv(sample_path)
```

```python
print(data.medium)
type(data.medium)
```

```python
# For series, we can use a class of methods on it which are called string methods
data.medium.str
```

```python
data.medium.str.contains('Graphite')
```

```python
# data.loc[rows, cols]
data.loc[data.medium.str.contains('Graphite'),['artist']] #all rows where string in medium column contains 'Graopite' 
```

```python
# string contains method is case sensitive by default, add case=False to ignore case sensitivity
data.loc[data.medium.str.contains('Graphite', case=False),['artist','medium']]
```

```python
data.loc[data.medium.str.contains('(?i)Graphite',regex=True), ['artist','medium']]
data.loc[data.medium.str.contains('(?i)Graphite',regex=True) | data.medium.str.contains('(?i)Line',regex=True), ['artist','medium']]
```

```python
data.loc[data.medium.str.contains('Graphite', case=False) | data.medium.str.contains('Line', case=False),['artist','medium']]
```

```python
# There is an easier way!
data.loc[data.medium.str.contains('graphite|line', case=False),['artist','medium']]
```

```python
# If you want to run string method on a column, make sure you convert it to a string first
data.year.astype(str).str.contains('1826')
data.loc[data.year.astype(str).str.contains('1826')]
```
